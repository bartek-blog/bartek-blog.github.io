```python
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
matplotlib.rcParams['figure.figsize'] = (20, 10)
```

```python
data = pd.read_csv("data/predict_future_sales/sales_train.csv.gz")
```

```python
data['dt'] = pd.to_datetime(data["date"], format="%d.%m.%Y")
data['sales'] = data['item_price'] * data['item_cnt_day']
```

```python
total_sales = data.groupby("dt", as_index=False)[['sales']].sum()
total_sales.set_index('dt').plot()
```
